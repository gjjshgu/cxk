```python
from Apriori import *
import pandas as pd
```

## 读取数据集

```python
df = pd.read_csv('mushrooms.csv')
df
```

## 简单的预处理
可以发现特征标记有很多重复（好多个w），导致难以区分是哪一个特征的w，因此在数据集前面加上类型，结果如下图

```python
col = df.columns.tolist()
data = df.values
for x in range(len(data)):
    for y in range(len(col)):
        data[x][y] = col[y] + '=' + data[x][y]
data[0]
```

## 训练
设支持度阈值为0.46，置信度阈值默认0.7

```python
ap = Apriori(min_support = 0.46)
ap.get_frequent_element(data)
res = ap.generate()
```

找出与毒蘑菇有关的特征

```python
for i in res:
    for j in range(2):
        if 'class=p' in i[j]:
            print(i[0], "--->", i[1], "conf", i[2])
```

## 修改数值
如果把支持度设为0.5会怎样呢？可以发现没有任何输出，说明毒蘑菇支持度小于0.5，被刷掉了

```python
ap = Apriori(min_support = 0.5)
ap.get_frequent_element(data)
res = ap.generate()
for i in res:
    for j in range(2):
        if 'class=p' in i[j]:
            print(i[0], "--->", i[1], "conf", i[2])
```

尝试提高置信度为0.8

```python
ap = Apriori(min_support=0.46, min_conf=0.8)
ap.get_frequent_element(data)
res = ap.generate()
for i in res:
    for j in range(2):
        if 'class=p' in i[j]:
            print(i[0], "--->", i[1], "conf", i[2])
```

如果把支持度设为0.48，置信度设为0.8，结果相对少很多

```python
ap = Apriori(min_support=0.48, min_conf=0.8)
ap.get_frequent_element(data)
res = ap.generate()
for i in res:
    for j in range(2):
        if 'class=p' in i[j]:
            print(i[0], "--->", i[1], "conf", i[2])
```
